```python
import numpy as np # calculo numerico
from IPython.display import HTML # para escribir HTML (ocultar la pagina)
```

```python
# Para hacer el plot 3D dinamico con cargar estos paquetes es suficiente
import plotly.plotly as py
import plotly.graph_objs as go
import pandas as pd
```

<h1><center>Cargar desde excel unos datos y hacer un plot</center></h1>

```python
# Cargar una hoja de datos
# xl = pd.ExcelFile("Test_001.xlsx")
data_file = pd.read_excel(open('C:\Dropbox (MGEP)\Ikerketa\PDSA\Anaconda_Jupyter\Test_002.xlsx','rb'), sheet_name='Hoja1')
```

```python
# Hacer algunas operaciones
A=data_file
B=np.transpose(A) # transpuesta
Z=np.outer(A,B) # multiplicar un vector por su transpuesta
```

```python
# Usar Plotly: para eso es necesario logearse
py.sign_in('Ekaitz', '8shIzm9a8J9BsgybCt6Q')
# Estructurar los datos para dibujar una superficie
data=[go.Surface(z=Z)] 
```

```python
# Todas las opciones de la figura
layout = go.Layout(
    title='Animated 3D chart', # titulo de la figura y los x,y,z labels
                        scene = dict(
                    xaxis = dict(
                        title='Prize'),
                    yaxis = dict(
                        title='Time'),
                    zaxis = dict(
                        title='Risk'),),
    autosize=False, # se controla el tamaño de la figura con esto
    width=800, # se controla el tamaño de la figura con esto
    height=800, # se controla el tamaño de la figura con esto
    margin=dict(
        l=65,
        r=50,
        b=65,
        t=90
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='Prueba_3D')
```
