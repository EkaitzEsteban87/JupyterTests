```python
import ipyvolume.pylab as p3
import numpy as np

from IPython.display import HTML

from matplotlib import cm

import pandas as pd
```

<h1><center>Cargar desde excel unos datos y hacer un plot</center></h1>

```python
# Cargar una hoja de datos
# xl = pd.ExcelFile("Test_001.xlsx")
data_file = pd.read_excel(open('C:\Dropbox (MGEP)\Ikerketa\PDSA\Anaconda_Jupyter\Test_002.xlsx','rb'), sheet_name='Hoja1')
```

```python
# Hacer algunas operaciones
A=data_file
B=np.transpose(A) # transpuesta
Z=np.outer(A,B) # multiplicar un vector por su transpuesta
```

```python
# Dibujar una superficie
X=np.linspace(1,20,Z.shape[1])
Y=np.linspace(12,40,Z.shape[1])
X, Y = np.meshgrid(X, Y)
```

```python
# Hacer el dibujo
colormap = cm.coolwarm
znorm = Z - Z.min()
znorm = znorm/znorm.max()
znorm.min(), znorm.max()
color = colormap(znorm)
p3.figure()
mesh = p3.plot_surface(X, Y, Z, color=color[...,0:3])
labelx="Time"
labely="Prize"
labelz="Risk"
p3.xyzlabel(labelx, labely, labelz)
p3.show()
```
